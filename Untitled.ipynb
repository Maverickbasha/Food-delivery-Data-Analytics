```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Zomato-data-.csv")
df.head()
```

```python
df = df.dropna()
```

```python
print(df.columns)
```

```python
df['listed_in(type)'].value_counts().plot(kind='bar')
plt.title("Popular Food Types")
plt.show()
```

```python
df['online_order'].value_counts().plot(kind='bar')
plt.title("Online Order Availability")
plt.show()
```

```python
df['approx_cost(for two people)'] = df['approx_cost(for two people)'].astype(str).str.replace(',', '')
df['approx_cost(for two people)'] = pd.to_numeric(df['approx_cost(for two people)'], errors='coerce')

sns.histplot(df['approx_cost(for two people)'])
plt.title("Cost Distribution")
plt.show()
```

```python
df['rate'] = df['rate'].replace('NEW', None)
df['rate'] = df['rate'].replace('-', None)
df['rate'] = df['rate'].astype(str).str.split('/').str[0]
df['rate'] = pd.to_numeric(df['rate'], errors='coerce')

sns.histplot(df['rate'])
plt.title("Ratings Distribution")
plt.show()
```
